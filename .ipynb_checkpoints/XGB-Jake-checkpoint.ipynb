```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt

from numpy import loadtxt
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from random import randint
from datetime import datetime
```

```python
train_path = 'data/signup_train_data.csv'
test_path = 'data/signup_test_data.csv'
pd.options.display.max_columns = 2000


# Load the data into a DataFrame 
df = pd.read_csv(train_path, low_memory=False)
test_df = pd.read_csv(test_path, low_memory=False)

# Split into X and Y
dfLabels = df.pop("label")
#print (df.head())

#Feature Engineering
#for index, row in df.iterrows():
#    df['time_of_day_int'] = introw['created_at'][11:13])*3600 + int(row['created_at'][14:16])*60 + int(row['created_at'][17:19])

df['employee_count_low'] = 0
df['employee_count_high'] = 0
df.loc[df['employee_count']  == '1 - 500', 'employee_count_low'] = 1
df.loc[df['employee_count']  == '1 - 500', 'employee_count_high'] = 500
df.loc[df['employee_count']  == '501 - 1,000', 'employee_count_low'] = 501
df.loc[df['employee_count']  == '501 - 1,000', 'employee_count_high'] = 1000
df.loc[df['employee_count']  == '1,001 - 5,000', 'employee_count_low'] = 1001
df.loc[df['employee_count']  == '1,001 - 5,000', 'employee_count_high'] = 5000
df.loc[df['employee_count']  == '5,000+', 'employee_count_low'] = 5001
df.loc[df['employee_count']  == '5,000+', 'employee_count_high'] = 10000
df['volume_high'] = 0
df.loc[df['employee_count']  == '1 - 100,000', 'volume_high'] = 100000
df.loc[df['employee_count']  == '100,001 - 2,500,000', 'volume_high'] = 2500000
df.loc[df['employee_count']  == '2,500,001 - 10,000,000', 'volume_high'] = 10000000
df.loc[df['employee_count']  == '10,000,000+', 'volume_high'] = 20000000
df.loc[df['employee_count']  == '0 to 40k', 'volume_high'] = 40000
df.loc[df['employee_count']  == '40k to 100k', 'volume_high'] = 100000

test_df['employee_count_low'] = 0
test_df['employee_count_high'] = 0
test_df.loc[df['employee_count']  == '1 - 500', 'employee_count_low'] = 1
test_df.loc[df['employee_count']  == '1 - 500', 'employee_count_high'] = 500
test_df.loc[df['employee_count']  == '501 - 1,000', 'employee_count_low'] = 501
test_df.loc[df['employee_count']  == '501 - 1,000', 'employee_count_high'] = 1000
test_df.loc[df['employee_count']  == '1,001 - 5,000', 'employee_count_low'] = 1001
test_df.loc[df['employee_count']  == '1,001 - 5,000', 'employee_count_high'] = 5000
test_df.loc[df['employee_count']  == '5,000+', 'employee_count_low'] = 5001
test_df.loc[df['employee_count']  == '5,000+', 'employee_count_high'] = 10000
test_df['volume_high'] = 0
test_df.loc[df['employee_count']  == '1 - 100,000', 'volume_high'] = 100000
test_df.loc[df['employee_count']  == '100,001 - 2,500,000', 'volume_high'] = 2500000
test_df.loc[df['employee_count']  == '2,500,001 - 10,000,000', 'volume_high'] = 10000000
test_df.loc[df['employee_count']  == '10,000,000+', 'volume_high'] = 20000000
test_df.loc[df['employee_count']  == '0 to 40k', 'volume_high'] = 40000
test_df.loc[df['employee_count']  == '40k to 100k', 'volume_high'] = 100000


#Temporarily parse non numerical data
frames_to_remove = ['zip', 'city', 'website', 'state', 'country', 'registration_ip', 'company','multifactor_country_code','created_at','lead_source','marketing_channel','user_persona','initial_package','geolocation_notes','name_notes','ip_notes','community_notes','email_notes','activity_notes','fingerprint_notes','employee_count','volume']
df = df.drop(frames_to_remove, axis=1)
test_df = test_df.drop(frames_to_remove, axis=1)

df.fillna(value=0, inplace=True)
test_df.fillna(value=0, inplace=True)

df.head()


```

```python
print(df.volume.unique())
```

```python
def xgb_basic(seed_in, TO_TEST, eta=0.1, depth=3, n_estimators=100, split=0.5):
    # split data into train and test sets
    seed = seed_in
    X_train, X_test, y_train, y_test = train_test_split(df, dfLabels, test_size=1-split, random_state=seed)

    # fit model no training data
    model = XGBClassifier(learning_rate=eta, max_depth=depth, n_estimators=n_estimators)
    model.fit(X_train, y_train)
    
    # make predictions for test data
    y_pred = model.predict(X_test)
    predictions = [round(value) for value in y_pred]

    # evaluate predictions
    accuracy = accuracy_score(y_test, predictions)
    print("Seed: %i Eta: %f Depth: %i Estimators: %i  -  Accuracy: %.3f%%" % (seed, eta, depth, n_estimators, accuracy * 100.0))
    
    if TO_TEST:
        #make predictions for train data
        y_pred = model.predict(test_df)
        predictions = [round(value) for value in y_pred]

        d = {'id':test_df['id'], 'label':predictions}
        preds = pd.DataFrame(data=d)

        preds.to_csv('predictions.csv', index=False)
```

```python
for split in [0.5,0.6,0.7,0.75,0.8,0.85,0.9,0.95]:
    xgb_basic(31,False, depth=9, eta=0.15, n_estimators=n_estimators, split=split)
```
